```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
x=dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
y_ac=ac.fit_predict(x)
print(y_ac)
```

```python
plt.title('Hierarchial Clustering is Easy')
plt.scatter(x[y_ac==0,0],x[y_ac==0,1],color='red',label='Cluster 1',s = 100)
plt.scatter(x[y_ac==1,0],x[y_ac==1,1],color='blue',label='Cluster 2',s = 100)
plt.scatter(x[y_ac==2,0],x[y_ac==2,1],color='magenta',label='Cluster 3',s = 100)
plt.scatter(x[y_ac==3,0],x[y_ac==3,1],color='cyan',label='Cluster 4',s = 100)
plt.scatter(x[y_ac==4,0],x[y_ac==4,1],color='green',label='Cluster 5',s = 100)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
